```python
file = open("data/金瓶梅.jieba.after_stop_words.txt", 'r')
text = file.read()
file.close()

text_list = text.split(" ")
```

```python
file = open("data/金瓶梅人物.dict", 'r')
characters = file.read()
file.close()

character_list = characters.split("\n")
```

```python
import pandas as pd

a = pd.DataFrame(text_list,columns=['人物'])
c = a.groupby('人物')['人物'].agg({'出场次数':'count'}).reset_index()
d = c.sort_values('出场次数',ascending=False)

character = pd.DataFrame()
character = d[d['人物'].isin(character_list)]
character.sort_values(by='出场次数',ascending=False, inplace=True) # 这个阶段排序可做可不做，可以等到查询结果后再排序
character.reset_index(drop=True, inplace=True) # reset index 可做可不做，reset 之后看起来整洁顺眼一点
```
